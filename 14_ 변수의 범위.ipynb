# **1. scope**
- 변수에 접근할 수 있는 범위
- local : 가장 가까운 함수 안 범위
- global : 함수 바깥의 변수 또는 import 된 module

```python
num1 = 10

def func1():
  num2 = 20
  print(num2)
```

```python
print(num1)
# print(num2) num2 라는 이름을 가진 변수가 없대요!
```

```python
# locals() : 로컬 변수를 확인하는 함수
# globals() : 글로벌 변수를 확인하는 함수

num1 = 10

def func1():
  num2 = 20
  print(num2)
  # 여기에서 num1 을 따옴표 안에 적어준 이유는
  # 따옴표 없이 적으면 저장된 값으로 인식하기 때문에
  # 변수 이름으로 인식하라고 따옴표를 씌워준다.
  print('num1이 로컬 변수가 맞아? : ', 'num1' in locals())
```

```python
func1() # num1 은 로컬 변수가 아니다.
```

```python
num1 = 10

def func1():
  num2 = 20
  print(num2)
  print('num1이 로컬 변수가 맞아? : ', 'num1' in globals())
```

```python
func1()
```

```python
num1 = 10

def func2():
  num1 = 20 # 로컬변수 num1 을 만들어서 20을 저장했을 뿐
  print(num1) # 전역변수 num1 을 바꿔준 건 아니다.
```

```python
func2()
print(num1)
```

# **2. global 키워드**
- 함수 내부에서 로컬 변수가 아닌 글로벌 변수로 사용하게 함

```python
wallet = 10000

def buy_juice():
  # 전역 변수 wallet 을 끌어다 사용하지 못하고 있다.
  wallet -= 1500 # wallet = wallet - 1500
```

```python
print(wallet)
# buy_juice()
print(wallet)
```

```python
wallet = 10000

def buy_juice():
  global wallet # 이 코드를 실행한 이 후부터 wallet 은 global 변수로 사용된다.
  wallet -= 1500
```

```python
print(wallet)
buy_juice()
print(wallet)
```
