# **2. 조건문**

### 2-1. if문

````python
# 비교(관계) 연산자
# 연산된 결과가 True, False 로 나온다.
# <, >, <=, >=, ==, !=
print(10 > 5)

# 야! 컴퓨터야! 10이랑 10이랑 같니?
# 네!(True) 같아요.
print(10 == 10)

# 야! 컴퓨터야! 10이랑 10이랑 다르니?
# 아니요!(False) 같아요.
print(10 != 10)
````

- if문 기본 문법
```
if 조건식(결과가 True 또는 False가 나오는 문장):
    조건식이 True인 경우 실행할 문장
```

````python
# 비교연산자를 사용해서 조건식을 만들 수 있다.

# 들여쓰기 신경쓰기!
age = int(input('나이를 입력하세요 : '))

if age > 19 :
  print('술 가져와')
  print('나 성인이야!')

# 조건식과 상관없이 무조건 실행되는 코드
print('입력한 나이 : ', age)
````

````python
age = int(input('나이를 입력하세요 : '))

# 실행할 문장이 하나 밖에 없을 때는 들여쓰기 없이 바로 이어써도 된다.
if age > 19 : print('술 가져와')

# 문법 오류 나는 코드!
# if age > 19 : print('술 가져와')print('헤헤')

print('입력한 나이 : ', age)
````

### 2-2. if ~ else 문
```
if 조건식(결과가 True 또는 False가 나오는 문장):
    조건식이 True인 경우 실행할 문장
else :
    조건식이 False인 경우 실행할 문장
```

````python
# 들여쓰기 신경쓰기!
age = int(input('나이를 입력하세요 : '))

if age > 19 :
  print('술 가져와')
  print('나 성인이야!')
else :
  print('나가')

# 조건식과 상관없이 무조건 실행되는 코드
print('입력한 나이 : ', age)
````

### 2-3. if ~ elif ~ else 문
```
if 조건식1(결과가 True 또는 False가 나오는 문장):
    조건식1이 True인 경우 실행할 문장
elif 조건식2 :
    조건식2가 True인 경우 실행할 문장
elif 조건식3 :
    조건식3가 True인 경우 실행할 문장
...
else :
    조건식이 False인 경우 실행할 문장
```
- elif 에 적은 조건식은 위에 조건식이 False 일 때만 검사한다.
- 즉, 위에 조건식이 True 라면 검사조차 하지 않는다.
- 위에 조건식의 결과에 상관없이 전부 다 검사하고 싶다면 elif 대신 if를 써야한다.

````python
age = int(input('나이를 입력하세요 >> '))

if age > 19 :
  print('성인입니다!')
elif age > 14 :
  print('청소년입니다!')
elif age > 6 :
  print('어린이입니다!')
elif age > 0 :
  print('유아입니다!')
else :
  print('생각이란 게 있나요~??')

print(f'입력한 나이 : {age}')
````

````python
# elif 를 사용하지 않고 if 만 사용해보기
# 위에 조건식의 여부와 상관없이 전부 검사!
# 전부 독립적인 if문으로 보기!
age = int(input('나이를 입력하세요 >> '))

if age > 19 :
  print('성인입니다!')
if age > 14 :
  print('청소년입니다!')
if age > 6 :
  print('어린이입니다!')
if age > 0 :
  print('유아입니다!')
else :
  print('생각이란 게 있나요~??')

print(f'입력한 나이 : {age}')
````

### 2-4. 삼항연산자 (참고)
- (True일 때 코드) if 조건식 else (False일 때 코드)

````python
# 삼항연산자는 조건식이 하나만 있을 때 자주 사용한다.
# 코드를 한 줄 만에 끝낼 수 있기 때문이다.
# 삼항연산자도 연산자이기 때문에 연산이 실행된 결과를 하나의 값으로 볼 수 있다.
num1 = int(input('숫자를 입력하세요 >> '))

print('5랑 크거나 같은 수를 입력했군요!') if num1 >= 5 else print('5보다 작은 수를 입력했군요!')
result = True if num1 >= 5 else False
print(result)
````

### 문제
- 국어점수, 영어점수, 수학점수를 입력 받아 총점, 평균, 학점을 구하는 프로그램을 작성해보자.
```
단, 학점은 평균 기준이며
90점 이상은 A학점,
80점 이상은 B학점,
70점 이상은 C학점,
60점 이상은 D학점,
나머지는 F학점으로 한다.
```

````python
kor, eng, math = input('국어, 영어, 수학점수를 스페이스바를 기준으로 입력하세요 >> ').split(' ')

total = int(kor) + int(eng) + int(math)
avg = total / 3

if avg >= 90 :
  grade = 'A'
elif avg >= 80 :
  grade = 'B'
elif avg >= 70 :
  grade = 'C'
elif avg >= 60 :
  grade = 'D'
else :
  grade = 'F'

print(f'총점 : {total}\n평균 : {avg:.2f}\n학점 : {grade}')
````

````python
# map(함수, 리스트)
# 리스트 하나하나의 요소에 전달한 함수 하나하나를 적용시킨다.
kor, eng, math = map(int, input('국어, 영어, 수학점수를 스페이스바를 기준으로 입력하세요 >> ').split(' '))

total = kor + eng + math
avg = total / 3

if avg >= 90 :
  grade = 'A'
elif avg >= 80 :
  grade = 'B'
elif avg >= 70 :
  grade = 'C'
elif avg >= 60 :
  grade = 'D'
else :
  grade = 'F'

print(f'총점 : {total}\n평균 : {avg:.2f}\n학점 : {grade}')
````

````python
# split 하면 각각을 구분해서 리스트로 반환
data = input('국어, 영어, 수학점수를 스페이스바를 기준으로 입력하세요 >> ').split(' ')
print(data)
````

### 문제 2.
- 숫자를 입력 받아 2의 배수인지, 3의 배수인지, 아니면 2, 3의 배수가 모두 아닌지를 판별하는 프로그램을 작성해보자.

````python
num = int(input('숫자를 입력!'))

if num % 6 == 0:
  print(f'{num}은(는) 2와 3의 공배수')
elif num % 2 == 0:
  print(f'{num}은(는) 2의 배수')
elif num % 3 == 0:
  print(f'{num}은(는) 3의 배수')
else :
  print(f'{num}은(는) 둘 다 아닙니다~!!')
````

````python
# 삼항연산자 문법을 활용해서 한 줄에 다 적어보자!

num = int(input('숫자를 입력!'))

print(f'{num}은 2와 3의 공배수') if num % 6 == 0 else print(f'{num}은 2의 배수') if num % 2 == 0 else print(f'{num}은 3의 배수') if num % 3 == 0 else print(f'{num}은(는) 둘 다 아닙니다~!!')
````

````python
# 파이썬에서는 삼항연산자를 사용해서 내가 실행하고 싶은 코드를 설정할 수 있다.
# 실행 코드 자리에 값을 적어놓는다면, 삼항연산자가 실행된 결과를 값으로 볼 수 있다.
result = 6 if num % 6 == 0 else 2 if num % 2 == 0 else 3 if num % 3 == 0 else -1

print(result)
````

2-5. 논리 연산자(and)
- 양쪽 두 조건의 값이 모두 True인 경우 결과가 True가 되는 연산자

| A | B | 결과 |
|-----|-----|-----|
|True|True|True|
|True|False|False|
|False|True|False|
|False|False|False|


````python
# 관리자 로그인 프로그램 만들기
# 아이디 : admin, 비밀번호 : 1234
# 연산자 우선순위
# 최단산쉬관리삼대
user_id = input('아이디를 입력하세요 >> ')
user_pw = input('비밀번호를 입력하세요 >> ')

if user_id == 'admin' and user_pw == '1234' :
  print('관리자가 로그인 하였습니다!')
else :
  print('관리자 아이디 또는 비밀번호를 확인하세요!')
````

### **참고**
- 연산자 우선순위
  - 최우선 -> ( )
  - 단항 -> !(조건)
  - 산술 -> +, -, /, *, % ...
  - 쉬프트 -> 비트 << , >>
  - 관계 -> 비교, <, >, <=, >=, ==, != ...
  - 논리 -> and, or
  - 삼항
  - 대입 -> =
- 결합성
  - 같은 우선 순위의 연산자들은 왼쪽부터 연산이 이루어진다.

### 2-6. 논리 연산자(or)
- 양 쪽 두 식의 값이 하나라도 True인 경우 결과가 True가 되는 연산자

| A | B | 결과 |
|-----|-----|-----|
|True|True|True|
|True|False|True|
|False|True|True|
|False|False|False|


````python
# 주민등록번호에 의한 성별 판별
serialNo = input('주민등록번호 13자리를 -없이 입력하세요!')

if serialNo[6] == '1' or serialNo[6] == '3' :
  print('남자에요!')
else :
  print('여자에요!')
````
