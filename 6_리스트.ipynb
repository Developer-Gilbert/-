# **1. 리스트(list)**
- 여러가지 데이터타입의 데이터를 ,(콤마)로 구분하여 [ ]로 감싼 자료구조

```python
# 리스트 만드는 방법1
li1 = [1, 3, 5, 7, 9]
print(li1)

# 리스트 만드는 방법2
# list() 함수 사용
li2 = list([1, 3, 5, 7, 9]) # 매개변수에는 반복가능한 객체를 넣어주면 리스트로 만들어준다.
print(li2)

# 문자만 가지고 만들어 보기
li3 = ['류호근', '홍길동', '강감찬', '논개', '수원이']
print(li3)

# 여러가지 데이터타입 섞어서 만들어 보기
li4 = [1, 3.14, '호근이', True]
print(li4)

# 리스트도 하나의 타입이기 때문에 여러가지 타입을 담을 수 있는 리스트에 담을 수 있다.
li5 = [1, 2, '치킨', ['류리멘탈', '파이썬']]
print(li5)
```

```python
# 리스트도 타입이란 것을 확인!
print(type(li4))
```

# **2. 리스트 인덱싱**

```python
print(li1)
print()

# 인덱싱 하는 방식은 문자열과 똑같다.
print(li1[0])
print(li1[-1])

# 리스트 인덱싱한 결과는 하나의 독립적인 값으로 봐야한다.
# 그렇기 때문에 데이터타입에 맞게 연산도 가능하다.
print(li1[-1] + 100)
print(li1[-1] + li1[0])
```

```python
li2 = ['류호근', '수원이', '하모', ['😍', '🥰', '💩', '🐔']]
print(li2[0])

# 리스트 타입 전체가 하나의 값
print(li2[3])
print(li2[3][3])
```

```python
li3 = [1, 3, 5, ['류호근', '홍길동', '강감찬', ['날', '떠', '나', '지', '마']]]

# '나' 뽑기
print(li3[-1][-1][-3])
print(li3[3][3][2])
```

# **3. 리스트 슬라이싱**

```python
# 슬라이싱도 문자열 슬라이싱과 똑같다.
li4 = [10, 20, 30, 40, 50]
print(li4)
print(li4[0:3])
print(li4[-3:-1])

# li4 는 데이터가 담겨있는 것처럼 보이지만 실제로는 주소값만 가지고 있다.
# 그렇기 때문에 li5 = li4 를 하게 된다면 주소를 넣어주는 것이다.
# 그래서 두 변수 모두 같은 주소를 참조하고 있기 때문에 하나의 값만 바꿔도 둘 다 바뀐다.
# 자주 실수 하는 모양!
li5 = li4
li5[0] = 100
print(li5)
print(li4)

# 값을 복사하고 싶다면 함수를 사용해야한다!
li6 = li4.copy()
print(li6)
li6[0] = 555
print(li6)
print(li4)
```

```python
# 참고
# 얕은 복사와 깊은 복사
# 리스트 객체 안에 또 다른 리스트 객체가 있을 경우에는
# copy() 함수는 얕은 복사가 된다.
test = [10, 20, 30, [40, 50, 60]]
temp = test.copy()
temp[0] = 100
print(test)
print(temp)
temp[-1][-1] = 600
print(test)
print(temp)
print('test의 내부 리스트의 주소:', id(test[-1]))
print('temp의 내부 리스트의 주소:', id(temp[-1]))

# 리스트 객체 안에 또 다른 리스트 객체가 있을 경우
# 깊은 복사 하는 방법
import copy
test = [10, 20, 30, [40, 50, 60]]
temp = copy.deepcopy(test)
print('\n---------------deepcopy() 사용-------')
print('test의 내부 리스트의 주소:', id(test[-1]))
print('temp의 내부 리스트의 주소:', id(temp[-1]))
```

# **4. 리스트 연산자**

```python
li7 = [10, 20, 30]
li8 = [40, 50, 60]

# 리스트끼리의 덧셈 연산은 연결!
print(li7 + li8)
print(li8 + li7)

# print(li7 - li8) # TypeError: unsupported operand type(s) for -: 'list' and 'list'
# print(li7 / li8) # TypeError: unsupported operand type(s) for /: 'list' and 'list'
# print(li7 * li8) # TypeError: can't multiply sequence by non-int of type 'list'

# 리스트 연산 중 곱셈은 정수랑만 가능
print(li7 * 3)
```

# **5. 리스트의 수정/삭제**

```python
# 인덱싱을 활용한 수정
li9 = [10, 20, 30]
print(li9)
li9[1] = 555
print(li9)
```

```python
li10 = [10, 20, 30]

# 슬라이싱을 사용해서 값을 수정할 때
li10[1:2] = [55, 66, 77]
# 슬라이싱을 이용하여 리스트 타입으로 수정한 경우 리스트에 데이터가 들어간다.
print(li10)

# 인덱싱을 사용해서 값을 수정할 때
li10 = [10, 20, 30]
li10[1] = [55, 66, 77]
# 인덱싱을 이용하여 리스트 데이터를 추가한 경우 리스트 안에 리스트가 들어간다.
print(li10)

# 삭제
# 슬라이싱을 이용한 삭제
li10[1:3] = [] # 빈 리스트를 저장 -> 삭제
print(li10)

# 인덱싱을 이용하여 똑같이 삭제한다면?
li10 = [10, 20, 30]
li10[1] = []
print(li10)

# del 키워드를 활용한 삭제
del li10[1]
print(li10)

# del 과 슬라이싱의 콜라보
li10 = [10, 20, 30]
del li10[0:2]
print(li10)
```

# **6. 리스트 함수**

```python
li11 = [10, 20, 30]
print(li11)

# 값 추가하기
# li11[3] = 40 # IndexError

# append() : 리스트 마지막 자리에 데이터를 추가
li11.append(40)
print(li11)

# 리스트도 하나의 값이기 때문에 추가 가능
li11.append([50, 60])
print(li11)

# 여러개 추가하는 건 안된다.
# li11.append(70, 80) # 매개 변수를 하나만 가질 수 있다. 하나씩 추가 가능.
```

```python
li12 = [10, 20, 30, 40, 50]
print(li12)

# pop() : 리스트 마지막에 저장된 데이터를 반환하고 해당 데이터는 삭제
print(li12.pop())
print(li12)

# 매개변수로 인덱스를 전달하면 해당 인덱스의 데이터가 삭제.
temp = li12.pop(1)
print(li12)
print(temp)
```

```python
li13 = [10, 20, 30, 40, 50, 20, 40]

# remove() : 매개변수로 전달된 데이터를 제거.
# 같은 데이터가 여러개 있는 경우 처음 출현하는 데이터를 삭제
li13.remove(20)
print(li13)

li13.remove(20)
print(li13)

# 삭제할 데이터가 없다면 에러 발생
# li13.remove(20)
```

```python
li14 = [10, 20, 30]

# insert() : 리스트의 특정 위치 (인덱스)에 데이터를 삽입
li14.insert(1, 100)
print(li14)
```

```python
li15 = [10, 20, 30]

# index() : 리스트에 저장된 데이터가 있을 경우 위치를 인덱스로 반환(없으면 에러)
print(li15.index(30))
# print(li15.index(100)) # values
```

```python
li16 = [45, 88, 66, 4, 23, 47]

# reverse() : 리스트에 저장된 데이터를 반대 순서로 바꿔줌
li16.reverse()
print(li16)
```

```python
li17 = [10, 40, 30, 100, 90, 50]

# sort() : 리스트에 저장된 데이터를 오름차순, 내림차순으로 정렬
li17.sort() # 오름차순
print(li17)

li17.sort(reverse = True) # 내림차순
print(li17)

# 문자도 정렬이 가능하다.
# 소문자보다 대문자가 더
li18 = ['Apple', 'apple', 'Banana', 'banana','Orange', 'orange']
li18.sort(reverse =True)
print(li18)

li19 = ['류호근', '강동원', '장동건', '김도진', '디카프리오']
li19.sort()
print(li19)
```

```python
li18 = ['Apple', 'apple', 'Banana', 'banana','Orange', 'orange']

# sorted() 원본데이터는 수정하지 않고, 정렬된 리스트를 반환.

# print(sorted(li18))
print(sorted(li18, reverse=True))
print(li18) # 원래 있던 리스트는 수정되지 안흥ㅁ.

# 정렬된 리스트를 계속 사용하고 싶다면 저장공간(변수)에 저장한다.
result = sorted(li18, reverse=True)
print(result)
```

```python
li19 = [10, 20, 30, 50, 20, 40, 30, 20]

# count() : 리스트에 포함된 특정 데이터의 갯수를 반환
print(li19.count(20))
print(li19.count(555))
```

```python
li20 = [1, 2, 3]

# extend() : 리스트를 더함
li20.extend([40, 50])
print(li20)

li21 = [1, 2, 3]
# li21 + [40, 50]
li21 += [40, 50] # 복합대입연산자
print(li21)
```

```python
# len() : 요소의 갯수 반환
print(len(li21))
```
