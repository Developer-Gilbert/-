# **3. ë¦¬ìŠ¤íŠ¸ ìŠ¬ë¼ì´ì‹±**

```python
# ìŠ¬ë¼ì´ì‹±ë„ ë¬¸ìì—´ ìŠ¬ë¼ì´ì‹±ê³¼ ë˜‘ê°™ë‹¤.
li4 = [10, 20, 30, 40, 50]
print(li4)
print(li4[0:3])
print(li4[-3:-1])

# li4 ëŠ” ë°ì´í„°ê°€ ë‹´ê²¨ìˆëŠ” ê²ƒì²˜ëŸ¼ ë³´ì´ì§€ë§Œ ì‹¤ì œë¡œëŠ” ì£¼ì†Œê°’ë§Œ ê°€ì§€ê³  ìˆë‹¤.
# ê·¸ë ‡ê¸° ë•Œë¬¸ì— li5 = li4 ë¥¼ í•˜ê²Œ ëœë‹¤ë©´ ì£¼ì†Œë¥¼ ë„£ì–´ì£¼ëŠ” ê²ƒì´ë‹¤.
# ê·¸ë˜ì„œ ë‘ ë³€ìˆ˜ ëª¨ë‘ ê°™ì€ ì£¼ì†Œë¥¼ ì°¸ì¡°í•˜ê³  ìˆê¸° ë•Œë¬¸ì— í•˜ë‚˜ì˜ ê°’ë§Œ ë°”ê¿”ë„ ë‘˜ ë‹¤ ë°”ë€ë‹¤.
# ìì£¼ ì‹¤ìˆ˜ í•˜ëŠ” ëª¨ì–‘!
li5 = li4
li5[0] = 100
print(li5)
print(li4)

# ê°’ì„ ë³µì‚¬í•˜ê³  ì‹¶ë‹¤ë©´ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•´ì•¼í•œë‹¤!
li6 = li4.copy()
print(li6)
li6[0] = 555
print(li6)
print(li4)
```

```python
# ì°¸ê³ 
# ì–•ì€ ë³µì‚¬ì™€ ê¹Šì€ ë³µì‚¬
# ë¦¬ìŠ¤íŠ¸ ê°ì²´ ì•ˆì— ë˜ ë‹¤ë¥¸ ë¦¬ìŠ¤íŠ¸ ê°ì²´ê°€ ìˆì„ ê²½ìš°ì—ëŠ”
# copy() í•¨ìˆ˜ëŠ” ì–•ì€ ë³µì‚¬ê°€ ëœë‹¤.
test = [10, 20, 30, [40, 50, 60]]
temp = test.copy()
temp[0] = 100
print(test)
print(temp)
temp[-1][-1] = 600
print(test)
print(temp)
print('testì˜ ë‚´ë¶€ ë¦¬ìŠ¤íŠ¸ì˜ ì£¼ì†Œ:', id(test[-1]))
print('tempì˜ ë‚´ë¶€ ë¦¬ìŠ¤íŠ¸ì˜ ì£¼ì†Œ:', id(temp[-1]))

# ë¦¬ìŠ¤íŠ¸ ê°ì²´ ì•ˆì— ë˜ ë‹¤ë¥¸ ë¦¬ìŠ¤íŠ¸ ê°ì²´ê°€ ìˆì„ ê²½ìš°
# ê¹Šì€ ë³µì‚¬ í•˜ëŠ” ë°©ë²•
import copy
test = [10, 20, 30, [40, 50, 60]]
temp = copy.deepcopy(test)
print('\n---------------deepcopy() ì‚¬ìš©-------')
print('testì˜ ë‚´ë¶€ ë¦¬ìŠ¤íŠ¸ì˜ ì£¼ì†Œ:', id(test[-1]))
print('tempì˜ ë‚´ë¶€ ë¦¬ìŠ¤íŠ¸ì˜ ì£¼ì†Œ:', id(temp[-1]))
```
