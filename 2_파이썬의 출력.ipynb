# **1. print 함수**
- 매개변수로 지정된 데이터를 출력하고 줄을 변경

```python
print('Hello Python')
print(10)
print('10')
print("집에 가면 뭐 먹지...?") # 문자열 출력은 따옴표를 사용함 '' 또는  ""
```

```python
print('나는 너를', '사랑해')
print('나는 너를', end='\n') # end : print 출력 후 어떻게 처리할 지 여부에 대한 옵션
print('사랑해')
print('나는 너를', end=' ')
print('사랑해')
print('나는 너를', end=' 완전 ')
print('사랑해')
print('나는 너를', end='😍')
print('사랑해')
```

```python
# 연산이 모두 이루어진 결과를 값으로 봐야한다!
print(10 + 5) # 덧셈
print(10 - 5) # 뺄셈
print(10 * 5) # 곱셈
print(10 / 3) # 나눗셈
print(10 % 3) # 나머지 연산자
print(10 // 3) # 몫 연산자
print(2 ** 3) # 지수 연산자

print('10 + 5')

print('10' + 5) # TypeError
```

# **2. 출력 서식 지정하기**

1. %d(정수), %f(실수), %s(문자열)를 사용하여 출력 서식을 지정
2. 서식을 제외한 문자는 입력한 그대로 출력이 됨
3. 서식을 지정해 출력할 데이터가 2개 이상인 경우 (소괄호)로 데이터를 묶어야 함

```python
print('숫자 : 100')
print('숫자 : %d' % 100)
print('숫자 : %d' % 100.5)
```

```python
print('실수 : %f' % 100.5)
```

```python
print('문자열 : %s' % '류호근!')
print('문자열 : %s' % 100)
```

```python
print('10 + 5 = 15')
print('%d + %d = %d' % (10, 5, 10 + 5))
```

```python
# %d, 자릿수 설정
print('%5d' % 100)
print('%5d' % 10000)
print('%5d' % 10000000000)
print('%05d' % 100)
```

```python
# %f, 자릿수 설정
print('%08.2f' % 856.12)
print('%6.2f' % 856.1256) # 반올림
print('%.2f' % 856.1)
```

```python
# %s
print('%7s' % 'apple')
```

```python
# 5 / 3
print('%d / %d = %d' % (5, 3, 5/3))
print('%d / %d = %f' % (5, 3, 5/3))
print('%d / %d = %.2f' % (5, 3, 5/3))
```

# **3. format 함수와 출력 인덱스 사용**

```python
# 인덱스(index) : 0부터 순서대로 진행되는 점수
print('{} {}'.format('잡채', '먹고싶다'))
print('{0}가 너무 {1}'.format('잡채', '먹고싶다'))
print('{1}가 너무 {0}'.format('잡채', '먹고싶다'))
```

```python
print('{0:5s}님은 {1:03d}살 입니다.'.format('류호근', 22))
```
