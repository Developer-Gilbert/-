# **1. input 함수**

- 사용자에게 데이터를 입력 받을 수 있는 함수
- 데이터를 문자형으로 입력

```python
name = input()
print(name)
print(type(name))
```

```python
name = input('이름을 입력해보아요 >> ')
print('%s님! 이름이 참 멋지네요!' % name)
```

```python
# split() : 매개변수로 전달된 문자를 기준으로 데이터를 나눠서 저장
data = '류호근 22 176'
name, age, height = data.split(' ')
print(name, age, height)
```

```python
name, age, height = input('이름, 나이, 키를 스페이스 바를 기준으로 나눠서 입력하세요 >> ').split(' ')
print(f'{name:5s}님의 나이는 {age}살이며, 키는 {height}cm 입니다.')
```

```python
name, age, height = input('이름, 나이, 키를 /를 기준으로 나눠서 입력하세요 >> ').split('/')
print(f'{name:5s}님의 나이는 {age}살이며, 키는 {height}cm 입니다.')
```

```python
# int() : 문자열을 숫자(정수)타입으로 변환
num1 = '100'
num2 = 50
print(int(num1) + num2)

num3 = '200'
num4 = '10'
print(int(num3 + num4))
```

### 문제
- 국어, 영어, 수학 점수를 입력 받아 총점, 평균을 구하는 프로그램을 작성해보자

```python
kor, eng, math = input('국어, 영어, 수학점수를 스페이스 바를 기준으로 입력하세요 >> ').split(' ')

# 확인 한 번 해봄
# print(f'{kor}, {eng}, {math}')

# print(f'총점은 {int(kor) + int(eng) + int(math)}점 이고 평균은 {(int(kor) + int(eng) + int(math))/3}점 이다.')

total = int(kor) + int(eng) + int(math)
avg = (int(kor) + int(eng) + int(math))/3

print(f'총점은 {total}점 이고 평균은 {avg:.2f}점 이다.')
```
