### **2.문자열 인덱싱**

- 문자열에 번호를 매겨 특정 문자를 찾을 수 있는 기능
- index 번호는 자동으로 부여된다.

```python
str = '문자열에 번호를 매겨 특정 문자를 찾을 수 있는 기능'
print(str)
```

```python
# index : 자동으로 부여가 되는 방 번호
# index는 0 부터 시작한다.
print(str[0])
print(str[1])
print(str[4]) # 공백 문자
print(str[5])
print(str[-1])
```

### **3.문자열 슬라이싱**

문자열에서 원하는 부분을 추출하는 기능

```python
str = '문자열에서 원하는 부분을 추출하는 기능'
print(str)
```

```python
print(str[0:4])
print(str[6:13])
print(str[:13])
print(str[6:]) # 6부터 끝까지
```

### **4.문자열함수**

문자열과 관련된 여러가지 기능(함수)

```python
str = '문자열 함수는 문자열과 관련된 여러가지 기능'
```

```python
# count() : 문자열에서 매개변수로 전달된 문자열의 출현 횟수를 반환
print(str.count('문'))
print(str.count('관련된'))
```

```python
# find() : 매개변수로 전달된 문자열이 처음 나오는 위치를 인덱스로 반환
print(str.find('관'))
print(str.find('웤')) # -1은 인덱스 번호가 아니다.

#print(str.find('관련된')) # 쓰는 목적이 다르다.
```

```python
# index() : 매개변수로 전달된 문자가 처음 나오는 위치를 인덱스로 반환
print(str.index('자'))
#print(str.indx('의'))  #ValueError
```

```python
# replace() : 특정 문자열을 치환(바꾼다.)
# 변수 안에 값을 바꿔버리는 게 아닌, 바꾼 값을 반환해준다.
print(str)
print(str.replace('기능','함수'))

print(str)
```

```python
# upeer() : 문자열을 모두 대문자로 변환
# lower() : 문자열을 모두 소문자로 변환
# 이 두 친구도 변환한 후에 반환해준다.
str = '문자열 function는 문자열과 관련된 여러가지 FUNCTION'

print(str.upper())
print(str.lower())
```

```python
# lstrip() : 문자열의 왼쪽 공백을 제거
# rstrip() : 문자열의 오른쪽 공백을 제거
# strip() : 문자열의 양쪽 공백을 제거
# 제거한 후 반환!
str = '       hello       python      '
print(str)
print(str.lstrip())
print(str.rstrip())
print(str.strip())
print(str.replace(' ',''))
```

```python
# join() : 문자열에 특정 문자열을 삽입
# 삽입 후 반환한다.
str = '너의모든순간'
print(' '.join(str))
print('헤헤'.join(str))
```
