# **1. 람다 함수**
- 단일문으로 표현되는 익명함수
- 코드상에서 한 번만 사용되는 기능이 있을 때, 굳이 함수로 만들지 않고 일회성으로 만들어 사용
- 함수 또한 클래스를 통해 생성된 객체
- return 형

```python
# 람다를 사용하지 않은 일반 함수 선언
def square(x):
  return x ** 3
```

```python
print(square(5))
```

```python
# 클래스를 통해 만들어진 객체!
lambda x: x ** 3
```

```python
# 진짜 함수를 선언한 것인지 확인해보자!
# 익명함수에 이름을 붙여주는 행위!
# 여러분들이 살면서 절대 만나지 않을 모양입니다.
test = lambda x: x ** 3
print(test(5))
```

```python
# 함수 바로 사용 가능!
# 이런 형식으로도 잘 쓰이지 않고,
# 다른 함수와 같이 많이 쓰인다.
(lambda x: x ** 3)(5)
```

```python
# sorted() : 정렬하는 함수, key 속성은 어떤 방식으로 정렬을 할 것인지를 지정
# key 속성에는 대게 함수가 들어간다.
```

```python
# 매개변수로 넘어온 문자열의 양쪽 공백을 제거 후
# 길이를 리턴하는 함수!
def myData(str):
  return len(str.strip())
```

```python
print(myData('         hogeun               '))
```

```python
data = ['watermelon     ', '        apple', '     banana     ', '         melon123']
```

```python
# sorted() 사용 시 key 전달 전과 후를 비교해보자.
print(sorted(data))
print(sorted(data, key = myData)) # 함수의 사용이 아닌 적용이기 때문에 소괄호는 넘기지 않는다.
```

```python
# 람다 함수 사용
print(sorted(data, key = lambda x: len(x.strip())))
```

# **2. filter 함수**
- 람다가 유용하게 사용되는 대표적인 함수
- 특정 조건을 만족하는 요소만 남기고 필터링 시킴
- 필터링을 해준다라는 것은 요소의 갯수가 달라질 수 있다.

```python
num = [1, 2, 3, 4, 5, 6, 7, 10, 14, 15, 18, 20]

temp = list(filter(lambda n:n%2 == 0, num))
print(temp)
```

# **3. map 함수**
- 각 원소를 주어진 수식에 따라 변형하여 새로운 값을 반환
- 요소의 갯수는 차이가 없다.

```python
num = [1, 2, 3, 4, 5, 6, 7, 10, 14, 15, 18, 20]

temp = list(map(lambda n: n * 0.02, num))
print(temp)
```
