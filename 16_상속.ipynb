# **1. 상속**
- 기존의 정의해둔 클래스의 기능을 그대로 물려받음
- 기존 클래스에 기능 일부를 추가하거나, 변경하여 새로운 클래스를 정의함
- 코드를 재사용할 수 있게 됨
- 안정적이고 유지보수에 용이함
- 상속을 해주는 기존 클래스를 부모, parent, super, base 클래스라고 부름
- 상속을 받는 새로운 클래스는 자식, child, sub 클래스라고 부름

```python
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def eat(self, food):
    print(f'{self.name}은(는) {food}을(를) 먹는다.')

  def sleep(self, hour):
    print(f'{self.name}은(는) {hour}시간 잡니다.')
```

```python
animal = Animal('강아지', 5)
print('animal.name : ',animal.name)
animal.eat('영양식')
animal.sleep(8)
```

```python
class Dog(Animal): # Animal 클래스를 Dog 클래스에 상속
  pass
```

```python
# Dog 클래스를 객체화 했지만,
# Dog 클래스는 Animal를 상속 받았기 때문에
# Animal 에 있는 기능들을 사용할 수 있다.
jin = Dog('진순이', 1)
jin.eat('사랑')
jin.sleep(10)
```

# **2. 메소드 오버라이딩**
- 부모 클래스의 method를 재정의
- 자식 클래스의 인스턴스로 호출시, 재정의된 메소드가 호출됨

```python
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def eat(self, food):
    print(f'{self.name}은(는) {food}을(를) 먹는다.')

  def sleep(self, hour):
    print(f'{self.name}은(는) {hour}시간 잡니다.')
```

```python
class Dog(Animal):
  def eat(self, food): # 메소드 오버라이딩, 재정의
    print(f'{self.name}은(는) {food}을(를) 아주 맛있게 먹는다.')
```

```python
tomy = Dog('토미', 15)
tomy.eat('개껌') # 자식 클래스에서 재정의된 메소드가 실행된다.
tomy.sleep(8)
```

# **3. 다중 상속**

```python
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def eat(self, food):
    print(f'{self.name}은(는) {food}을(를) 먹는다.')

  def sleep(self, hour):
    print(f'{self.name}은(는) {hour}시간 잡니다.')
```

```python
class Human:
  def __init__(self, name, age, height):
    self.name = name
    self.age = age
    self.height = height

  def study(self, hour):
    print(f'{self.name}은(는) {hour}시간 동안 공부합니다.')
```

```python
# 생성자가 둘 다 있는 클래스를 상속 받을 경우에는
# 앞에 적어준 생성자가 우선순위가 더 높다.
# 즉, Ryu 클래스를 객체화 할 때 Human 클래스의 생성자를 사용하여 객체화하여야한다.
class Ryu(Human, Animal):
  # 다중상속 받는 자식에서 생성자를 재정의하면
  # Ryu 클래스를 이용해서 객체를 만들 때, 자식에서 재정의한 생성자가 호출된다.
  def __init__(self, name, age, height, weight):
    self.name = name
    self.age = age
    self.height = height
    self.weight = weight
```

```python
ryu = Ryu('류호근', 22, 175)
ryu.eat('잡채')
ryu.sleep(5)
ryu.study(6)
print(ryu.height)
```
