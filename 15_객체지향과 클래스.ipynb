### 1-3. 생성자
- 클래스 인스턴스가 생성될 때 자동으로 호출 됨
- __ init __(self, 변수1, 변수2, ...):
- self 매개변수는 항상 첫번째 오며, 자기 자신의 객체를 가리킴
- 이름이 꼭 self일 필요는 없지만 관례적으로 self를 사용
- 생성자에서는 해당 클래스가 다루는 데이터를 정의

````python
class Dog:
  def __init__(self): # 셀프는 객체 그 자체!
    print(self, 'init 호출')
````

````python
jin = Dog()
print(jin)

tomy = Dog()
print(tomy)
````

````python
class Dog:
  def __init__(self): # 셀프는 객체 그 자체!
    print(self, 'init 호출')
    self.name = '이름없음'
    self.age = 0
````

````python
jin = Dog()
print(jin.name)
print(jin.age)

jin.name = '진순이'
print(jin.name)

tomy = Dog()
print(tomy.name)
````

````python
class Dog:
  def __init__(self, name, age, family = '진돗개'): # 셀프는 객체 그 자체!
    print(self, 'init 호출')
    self.name = name
    self.age = age
    self.family = family
````

````python
jin = Dog('진순이', 1)
print(jin.name)
print(jin.family)

tomy = Dog('토미', 15, '시고르자브종')
print(tomy.name)
print(tomy.age)
print(tomy.family)
````

### 1-4. 메소드 정의하기
- 해당 클래스의 객체에서만 호출가능한 함수
- 해당 객체의 속성에 대한 연산을 행함
- 객체이름.메소드명() 형태로 호출함.

````python
class Counter:
  def __init__(self):
    self.num = 0
  # 번호표를 뽑으면 번호표 번호를 +1 해주는 기능
  def increment(self):
    self.num += 1
  # 현재 번호표 번호를 리턴하는 기능
  def get_current_value(self):
    return self.num
  # 번호표를 리셋해주는 기능
  def do_reset(self):
    self.num = 0
````

````python
SHBank = Counter()
KBBank = Counter()
print('신한 번호표 : ', SHBank.num)
print('국민 번호표 : ', KBBank.num)

# increment 한 객체는 신한은행 객체!
# 그렇기 때문에 국민은행 객체는 아무 영향을 받지 않는다.
SHBank.increment()
print('신한 현재 번호 : ', SHBank.get_current_value())
print('국민 현재 번호 : ', KBBank.get_current_value())
````

### 1-5. 메소드 타입
- instance method : 객체 형태로 호출되기 때문에 해당 메소드를 호출한 객체에만 영향을 미침
- class method : class 로 호출 (함수 선언 위에 @staticmethod 라고 표기)


````python
class Math:
  def add(self, x, y):
    return x + y
  def multiply(self, x, y):
    return x * y
````

````python
math = Math()
result1 = math.add(10, 3)
print(result1)
````

````python
class Math:
  # 어노테이션
  # 아래 영역의 쓰임을 컴퓨터에게 알려주는 용도.
  @staticmethod
  def add(x, y):
    return x + y

  @staticmethod
  def multiply(x, y):
    return x * y
````

````python
# 메소드의 소속만 알려주면 된다.
# 객체를 만들 필요 없이 바로 사용가능.
result2 = Math.add(40, 9)
print(result2)
````
