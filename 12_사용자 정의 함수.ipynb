### 1-2. 매개변수(parameter)가 있는 함수
```
def 함수명(변수1, 변수2, ...):
    함수가 호출되면 실행할 문장
    ...
```
```
함수명(값1, 값2, ...) # 호출
```

````python
# func2(17) 이 형태로 호출되었다면
# n = 17 이 실행이 된 이후
# 함수 내에 선언되어있는 코드가 실행이 된다.
def func2(n):
  print('입력받은 숫자 : %d' % n)
````

````python
num1 = int(input('숫자를 입력 >> '))

func2(num1)
func2(17)

# 매개변수를 설정해놨는데 넘겨주지 않는다면
# 에러가 발생한다.
# 발생하지 않는 경우도 있다. 그건 밑에 디폴트 매개변수 참고!
# func2()
````

````python
# 두 정수를 매개변수로 넘겨주고,
# start 부터 end 까지 정수의 합을 구하는 함수
def func3(start, end):
  sum = 0
  for i in range(start, end+1):
    sum += i
  print(f'{start}부터 {end}까지의 총 합은 {sum}입니다.')
````

````python
func3(50, 100)

# 이 모양은 잘 사용하지 않는다.
# 매개변수는 기본적으로 작성해준 순서대로 들어가게 된다.
func3(end = 10, start = 1)
````

### 1-4. 디폴트 매개변수
```
def 함수명(변수1=값1, 변수2=값2, ...):
    함수가 호출되면 실행할 문장
    ...
    return 값
```
```
변수 = 함수명(값1, 값2, ...) # 호출
변수 = 함수명() # 호출
```

````python
# 디폴트 매개변수를 사용하지 않은 함수
def func7(num1, num2):
  return num1 * num2
````

````python
print(func7(10, 5))

# 매개변수를 넘겨주지 않았기에,
# 매개변수가 초기화되지 않으면서 오류!
# print(func7())
````

````python
# 디폴트 매개변수를 사용한 함수
def func7(num1=4, num2=9):
  return num1 * num2
````

````python
# 매개변수를 넘겨주지 않으면
# 디폴트 매개변수로 넘겨준 값이 자동으로 들어가면서
# 함수는 정상적으로 호출!
print(func7())
````

### 1-5. 매개변수의 개수가 가변적인 함수
```
def 함수명(*변수):
    함수가 호출되면 실행할 문장
    ...
    return 값
```
```
변수 = 함수명(값1, 값2, ...) # 호출
변수 = 함수명() # 호출
```
> 매개변수는 함수 안에서 튜플형태로 사용

````python
def func8(*args):
  print(args) # 매개변수가 어떤 식으로 넘어왔는지 확인하기 위하여 프린트해봄
  print(type(args)) # 타입도 확인!
  result = 0
  for i in args:
    result += i
  return result
````

````python
print(func8(10))
print(func8(10,20,30,40,50))
````

### 1-6. 매개변수를 딕셔너리 구조로 전달받는 함수
```
def 함수명(**변수):
    함수가 호출되면 실행할 문장
    ...
    return 값
```
```
변수 = 함수명(키1=값1, 키2=값2, ...) # 호출
```

````python
def func9(**args):
  # print(args)
  # print(type(args))
  for i in args:
    if i == 'user_id':
      return args[i]
# return 함수의 실행 결과 값을 넘겨줄 때도 사용하고,
# 또 함수의 종료를 의미한다.
````

````python
userId = func9(user_id = 'hogeun', name = '류호근', age = 22)
print(userId)
````

### 1-7. 리턴값이 여러개인 함수
```
def 함수명(변수1, 변수2, ...):
    함수가 호출되면 실행할 문장
    ...
    return 값1, 값2, ...
```
```
변수1, 변수2, ... = 함수명(값1, 값2, ...) # 호출
```

````python
def func10(num1, num2):
  return num1+num2, num1-num2, num1*num2, num1/num2
````

````python
temp = func10(10, 3)
print(temp) # 여러값을 리턴을 하면, 그 값들이 튜플에 묶여서 넘어온다.
````

````python
result1, result2, result3, result4 = func10(10, 3)
print(result1, result2, result3, result4)
````

````python
_, _, result3, _ = func10(12, 3)
print(result3) # 곱셈의 결과만 변수에 담아주고 싶을 때
````

````python
# 곱셈의 결과만 필요할 때
temp2 = func10(12, 3)[2]
print(temp2)
````

### 문제
- 700원짜리 음료수를 뽑을 수 있는 자판기 프로그램을 구현하세요.
- 출력예시
```
음료수 = 0개, 잔돈 = 3000원
음료수 = 1개, 잔돈 = 2300원
음료수 = 2개, 잔돈 = 1600원
음료수 = 3개, 잔돈 = 900원
음료수 = 4개, 잔돈 = 200원
```

````python
def vending_machine(money):
  price = 700
  count = money // price
  for i in range(count+1):
    change = money - price * i
    print(f'음료수 = {i}, 잔돈 = {change}원')
````

````python
vending_machine(3000)
````

````python
def vending_machine(money):
  drink=0
  while money>=0:
    print(f'음료수 = {drink}개, 잔돈 = {money}원')
    money-=700
    drink+=1
````
