# **1. 딕셔너리(Dictionary)**
- 대응관계를 나타내는 자료형으로 key와 value라는 것을 한 쌍으로 갖는 형태
- 하나의 딕셔너리의 key는 중복될 수 없음
- 하나의 딕셔너리의 value는 중복될 수 있음

### 1-1. 딕셔너리 만들기

```python
dic1 = {}
print(type(dic1))
```

```python
dic2 = {1:'류호근', 2:'홍길동', 3:'강감찬', 4:'장발장'}
print(dic2)
```

### 1-2. key를 통해 value 찾기

```python
# 딕셔너리는 인덱스 번호가 존재하지 않는다.
# value 에 접근하기 위해서는 key 값을 알고 있어야한다.
dic2 = {1:'류호근', 2:'홍길동', 3:'강감찬', 4:'장발장'}
print(dic2[1])
print(dic2[3])
```

```python
# key 값을 같은 자료형으로 해줄 필요는 없다.
dic3 = {'no':1, 'user_id':'hogeun', 'name':'류호근', 4:22}
print(dic3['no'])
print(dic3['name'])
print(dic3[4])
```

### 1-3. 데이터 추가 및 삭제

```python
dic4 = {1:'apple'}
print(dic4)

# 딕셔너리에 데이터 추가하기
# 데이터를 수정하는 문법을 사용해서 추가도 할 수 있다.
# 해당 딕셔너리에 없는 키값을 저장공간으로서 사용한다면,
# 키 값을 하나 만들어서 데이터를 저장한다.
dic4[100] = 'orange'
print(dic4)

dic4['favorite'] = 'mandarin'
print(dic4)

# 수정
dic4['favorite'] = 'watermelon'
print(dic4)

# 삭제
del dic4[100]
print(dic4)
```

### 1-4. 딕셔너리 함수

```python
dic3 = {'no':1, 'user_id':'hogeun', 'name':'류호근', 4:22}

# keys() : key 리스트를 반환
print(dic3.keys())

# values() : value 리스트를 반환
print(dic3.values())

# items() : key와 value를 한 쌍으로 묶는 튜플을 반환
print(dic3.items())

print('----- get 부분 -----')

# get() : key를 이용해서 value를 반환
print(dic3['user_id'])
# print(dic3['age']) # 해당 딕셔너리에 키 값이 없는 친구를 찾는다면, 에러가 난다.
print(dic3.get('user_id'))
print(dic3.get('age')) # 없는 키값을 전달하면 에러가 아닌 None 을 반환
print(dic3.get('age', '먼저 나이를 입력하세요!'))

# in : key가 딕셔너리 안에 있는지 확인, 논리타입의 값을 반환
print('\n----- in -----')
print('name' in dic3) # True
print('age' in dic3) # False
```

### 1-5. 딕셔너리와 for

```python
dic3 = {'no':1, 'user_id':'hogeun', 'name':'류호근', 4:22}

for i in dic3:
  print(i, end = ' ') # 키만 가지고 나온다.

print('\n')

for i in dic3.keys():
  print(i, end = ' ') # 키만 가지고 나온다.

print('\n')

for i in dic3.values():
  print(i, end = ' ') # value 만 가지고 나온다.

print('\n')

for i in dic3:
  print(dic3[i], end = ' ')

print('\n')

for i in dic3:
  print(dic3.get(i), end = ' ')
```

### 문제
- dic3 에서 모든 key 중 value가 'hogeun'이 있는지 확인하는 프로그램을 작성해보자!
  - 지금 현재 hogeun 이 있기 때문에, 코드셀을 실행했을 때 OK! 있네요! 가 출력되도록 하기
  - dic3 를 수정해서 hogeun이 없다면 NO! 없네요! 출력되도록 하기

```python
dic3 = {'no':1, 'user_id':'hogeun', 'name':'류호근', 4:22}

for i in dic3:
  if dic3[i] == 'hogeun':
    print('OK! 있네요!')
  else:
    print('NO! 없네요!')
```

### 문제
아래와 같은 딕셔너리 자료에서 주민등록번호를 추출하여 성별을 알아내는 프로그램을 작성해보자.
> user = {"name":"류호근", "hp":"010-1111-0000", "ssn":"0208293148567"}

```python
user = {"name":"류호근", "hp":"010-1111-0000", "ssn":"0208293148567"}

ssn = user.get('ssn')
# print(ssn)

if 'ssn' in user:
  if ssn[6] == '1' or ssn[6] == '3':
    print('남자')
  elif ssn[6] == '2' or ssn[6] == '4':
    print('여자')
  else:
    print('아직... 태어나지 않을텐데...? 확인해봐용')
else:
  print('ssn이 없는데용')
```
